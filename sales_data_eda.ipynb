What was the best month for sales? How much was earned that month?

```python
all['Quantity Ordered'] = all['Quantity Ordered'].astype(int)
all['Price Each'] = all['Price Each'].astype(float)

all['Total Sales'] =all['Quantity Ordered']* all['Price Each']
```

```python
all
```

```python
sales_month=all.groupby('month')['Total Sales'].sum().reset_index()
sales_month
```

```python
sales_month.plot(kind='bar')

plt.show()
```

```python
sns.barplot(x='month', y='Total Sales', data=sales_month, palette='viridis')
```

Which city has the highest number of orders?

```python
all['city']=all['Purchase Address'].str.split(',').str.get(1)
```

```python
all
```

```python
city_highest=all['city'].value_counts().reset_index()
city_highest
```

```python
#city_highest.columns= ['City','Count']
plt.pie(city_highest['count'], labels=city_highest['city'], autopct='%1.1f%%', startangle=140)
```

```python
all['hour']=all['time'].str[:2]
```

```python
sales_by_hour=all.groupby(['hour'])['Total Sales'].count().reset_index()
```

```python
sales_by_hour
```

```python
sales_by_hour.plot(kind='bar',color='green')
plt.xlabel('Order hour')
plt.ylabel('Count of Total Sales')
```

what products are most often sold together?

```python
sold_together=all[all['Order ID'].duplicated(keep=False)]
```

```python
sold_together
```

```python
c=sold_together.groupby('Order ID').agg({'Product':lambda x:','.join(x)}).reset_index()
c
```

```python
c.set_index('Order ID')
c.head()
```

```python
product_counts = c['Product'].value_counts()
top5=product_counts.head(5)
```
